```python
#Z TESTİ
#Departmanlarda çalışanlara verilen ortalama asgari ücreti 90312 TL ve standart sapması 31240 TL'dir.
#Rastgele 500 çalışan incelenip ortalama asgari ücretin %5 anlamlılık düzeyinde popülasyonun ortalamasına eşit olup olmadığı test edildi.
# H0: μ = 90312
# H1: μ ≠ 90312
import math
from numpy.random import randn
from statsmodels.stats.weightstats import ztest


sh = 31240 / math.sqrt(500)
ort=102020

df = sh * randn(500) + ort

ztest_değeri, p_value = ztest(df, value=100000, alternative='two-sided')

alfa = 0.05

print("p_value :", p_value )
print("ztest_değeri :", ztest_değeri )

if p_value < alfa:
    print('Boş hipotez reddedilir.')
else:
    print('Boş hipotez reddedilemez.')
```

```python
#T TESTİ
#Veri setinde çalışanların ortalama asgari ücreti 90312 TL'dir. 10 kişiden oluşan bir örneklem alıp %5 anlamlılık düzeyinde ortalama asgeri maaşları test edilmiştir.
from scipy.stats import ttest_1samp

a= [58571.1182, 66572.0, 166140.03, 98349.0, 82146.3286, 75621.0, 49230.0, 24779.2732, 21862.0, 68473.5124]

ort = 90312

t_statistic, p_value = ttest_1samp(a, ort)


alpha = 0.05

if p_value < alpha:
    print("Boş hipotezi reddet")
else:
    print("Boş hipotezi reddedilemez")


print("t-statistic:", t_statistic)
print("p-value:", p_value)
```
